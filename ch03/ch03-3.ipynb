```python
img = cv2.imread('/content/gdrive/My Drive/Colabip/Images/lena256.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gimg1 = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg1.shape)
plt.imshow(gimg1)
```

```python
img = cv2.imread('/content/gdrive/My Drive/Colabip/Images/circle.bmp')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gimg2 = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg2.shape)
plt.imshow(gimg2)
```

```python
gBit = np.zeros((gimg1.shape[0],gimg1.shape[1]),dtype =np.ubyte)
```

```python
for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    bitVal = gimg1[h,w]&gimg2[h,w]
    # bitVal = gimg1[h,w] | gimg2[h,w]
    # bitVal = gimg1[h,w] ^ gimg2[h,w]
    gBit[h,w] =bitVal

cv2_imshow(gBit)
```

```python
for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    # bitVal = gimg1[h,w]&gimg2[h,w]
    bitVal = gimg1[h,w] | gimg2[h,w]
    # bitVal = gimg1[h,w] ^ gimg2[h,w]
    gBit[h,w] =bitVal

cv2_imshow(gBit)
```

```python
for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    #bitVal = gimg1[h,w]&gimg2[h,w]
    #bitVal = gimg1[h,w] | gimg2[h,w]
    bitVal = gimg1[h,w] ^ gimg2[h,w]
    gBit[h,w] =bitVal

cv2_imshow(gBit)
```
